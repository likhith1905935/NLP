```python
import requests
from bs4 import BeautifulSoup

print("Libraries 'requests' and 'BeautifulSoup' imported successfully.")
```

```python
url = "https://en.wikipedia.org/wiki/Global_warming"

try:
    response = requests.get(url)
    response.raise_for_status()
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    print(f"Successfully fetched and parsed content from {url}")
except requests.exceptions.RequestException as e:
    print(f"Error fetching content: {e}")
    soup = None
```

```python
url = "https://en.wikipedia.org/wiki/Global_warming"


headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

try:
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    html_content = response.text
    soup = BeautifulSoup(html_content, 'html.parser')
    print(f"Successfully fetched and parsed content from {url}")
except requests.exceptions.RequestException as e:
    print(f"Error fetching content: {e}")
    soup = None
```

```python
if soup:

    content_div = soup.find('div', class_='mw-parser-output')

    if content_div:

        for element in content_div.find_all(['table', 'div', 'sup', 'style', 'script', 'nav', 'figure'], recursive=True):
            element.extract()


        paragraphs = content_div.find_all('p', recursive=True)
        article_text = '\n'.join([p.get_text() for p in paragraphs])
        print("Successfully extracted article text. First 500 characters:\n")
        print(article_text[:500])
    else:
        article_text = "Could not find the main content div."
        print(article_text)
else:
    article_text = "Soup object is empty, cannot extract text."
    print(article_text)
```

```python
zqqszqwdedwimport re

if soup:

    content_div = soup.find('div', class_='mw-parser-output')

    if content_div:

        for script_or_style in content_div(['script', 'style']):
            script_or_style.extract()

        for element_to_remove in content_div(['sup', 'table', 'nav', 'figure']):
            element_to_remove.extract()


        unwanted_div_classes = [
            'infobox', 'reflist', 'portal', 'toc', 'navbox', 'metadata', 'hatnote',
            'mw-parser-output-disambiguation', 'mw-references-columns', 'printfooter',
            'catlinks', 'stub', 'mw-indicator', 'thumb', 'reference'
        ]
        for div_class in unwanted_div_classes:
            for div_tag in content_div.find_all('div', class_=div_class):
                div_tag.extract()


        article_text = content_div.get_text()


        article_text = re.sub(r'\n\n+', '\n', article_text)

        article_text_lines = [line.strip() for line in article_text.split('\n')]
        article_text = '\n'.join(article_text_lines)

        article_text = re.sub(r'\[\d+\]', '', article_text)

        article_text = re.sub(r'\s\s+', ' ', article_text)


        print("Successfully extracted and cleaned article text. First 1000 characters:\n")
        print(article_text[:1000])
    else:
        article_text = "Could not find the main content div."
        print(article_text)
else:
    article_text = "Soup object is empty, cannot extract text."
    print(article_text)
```
