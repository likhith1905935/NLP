```python
import nltk
from collections import Counter
import pandas as pd

# Ensure you have the necessary NLTK data
nltk.download('punkt')
nltk.download('punkt_tab')

corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?',
]

# Tokenize and lowercase
tokenized_docs = [nltk.word_tokenize(doc.lower()) for doc in corpus]

# Flatten to get all unique words (vocabulary) - Fixed iteration
vocabulary = sorted(list(set([word for doc in tokenized_docs for word in doc if word.isalnum()])))

# Generate frequency vectors
bow_data = []
for doc in tokenized_docs:
    counts = Counter(doc)
    row = [counts.get(word, 0) for word in vocabulary]
    bow_data.append(row)

# Display as DataFrame
nltk_bow_df = pd.DataFrame(bow_data, columns=vocabulary)

print("Vocabulary:", vocabulary)
display(nltk_bow_df)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pandas as pd

# Sample data
corpus = [
    'This is the first document.',
    'This document is the second document.',
    'And this is the third one.',
    'Is this the first document?',
]

# Initialize the CountVectorizer
vectorizer = CountVectorizer()

# Fit and transform the corpus
X = vectorizer.fit_transform(corpus)

# Convert to a DataFrame for better visualization
bow_df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out())

print("Vocabulary:", vectorizer.get_feature_names_out())
display(bow_df)
```
